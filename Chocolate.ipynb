```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\brand\OneDrive\Documents\Bootcamp Projects\Case Studies\31.3.2 Craft A Story from a Dataset\flavors_of_cacao.csv")
df.columns
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace('\n',' ')
df2 = df.dropna()
print(df2['Bean Type'].unique())
```

```python
df2
```

```python
sns.barplot(data=df2, x='Bean Type',y ='Rating',color='red')
plt.xticks(rotation=90)
plt.title('Rating by bean type')
plt.xlabel('Bean Type')
plt.ylabel('Rating')
plt.savefig('Rating_by_bean_type.png',format='png',dpi=300,bbox_inches='tight')
plt.tight_layout()
plt.clf()
top5_bean_types = df.groupby('Bean Type')['Rating'].mean().sort_values(ascending=False).head(5)
print('Top 5 bean types:\n',top5_bean_types)
```

```python
#Finding Country Origin for top 5 bean types
print('Top 5 Country Origin:\n')
for index,row in df.iterrows():
    if row['Bean Type'] in top5_bean_types:
        print (row['Broad Bean Origin'])
```

```python
#Finding Company Location for top 5 bean types
print('Top 5 Company Country Origin:\n')
for index,row in df.iterrows():
    if row['Bean Type'] in top5_bean_types:
        print (row['Company Location'])
```
